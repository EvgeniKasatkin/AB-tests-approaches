```python
import seaborn as sns 
import warnings 
import statistics
from statsmodels.stats.proportion import proportions_ztest
import pandas as pd 
import numpy as np
warnings.filterwarnings('ignore')
```

```python
print('Input baseline conversion')
baseline_conversion = input()
print('Input sample size')
sample_size = input()
```

```python
#Monte-Carlo simulations:
number_of_simulations = 10000
result = []

for ind_ in range(number_of_simulations):
    np.random.seed(ind_)
    result.append(np.random.binomial(int(sample_size), float(baseline_conversion)))

result_df = pd.DataFrame({'conversions': result})
result_df['deviation'] = result_df['conversions'] - int(sample_size) * float(baseline_conversion)
result_df
```

```python
sns_plot = sns.distplot(result_df['deviation'])
fig = sns_plot.get_figure()
```

```python
#p-value distribution:

variance = statistics.variance(list(result_df['deviation']), xbar=None)
sigma = np.sqrt(number_of_simulations*variance/(number_of_simulations -1))
sigma = round(sigma) 

df_for_pvalue = pd.DataFrame({'Possible Conversions': list(range(round(int(sample_size) * float(baseline_conversion)) - 3*sigma, 
                                                       round(int(sample_size) * float(baseline_conversion)) + 3*sigma))})

df_for_pvalue['p_value_ztest'] = df_for_pvalue['Possible Conversions'].apply(lambda x: proportions_ztest(x, int(sample_size), float(baseline_conversion), alternative = 'two-sided')[1])
df_for_pvalue = df_for_pvalue[(df_for_pvalue['p_value_ztest'] <= 0.05) & (df_for_pvalue['p_value_ztest'] >= 0.03)]
df_for_pvalue['Baseline_conversion'] = round(int(sample_size) * float(baseline_conversion))
df_for_pvalue
```
